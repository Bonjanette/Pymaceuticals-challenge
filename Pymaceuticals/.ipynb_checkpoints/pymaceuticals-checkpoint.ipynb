### Analysis

> <u>Observation 1:</u> There is a strong correlation between mouse weight and the average observed tumor volume for the mice in the Capomulin regimen as the Pearson's Correlation is .84. From the Linear regression we see that the correlation is positive meaning that the greater the mouse weight, the greater the average tumor volume. It would be interesting to make the same plot for each treatment regimen to see if the regimens appear to have an affect, or if this is just typical of mice with these tumors.

> <u>Observation 2:</u> Capomulin and Ramicane have very similar Boxplots, and Ramicane has a slightly lower average final Tumor Volume. Investigation of Ramicane might be warranted as the average final tumor volume for that regimen is slightly smaller and therefore potentially a better outcome. What this analysis doesn't show is the percentage of shrinkage of the tumors. This is important because without considering the initial tumor volumes, one cannot determine the magnitude of the affect of regimens as it is possible that one group of mice began with tumors of a different average size. A measure of the change of tumor size for each regimen would be potentially insightful.

> <u>One could infer</u> from the mean and median values listed in the Summary Statistics table that the only two regimens in the study warranting further investigation are those of Capomulin and Ramicane as they have significantly lower values than the other regimens considered. In fact the other regimens have similar values in the summary to the Placebo tumor volumes, which means they appear to have had no positive impact on the tumor volumes in this study.


## Quartiles, Outliers and Boxplots


```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
 
# Start by getting the last (greatest) timepoint for each mouse
greatestTimepoints = mouse_study_reduced_df.groupby('Mouse ID')['Timepoint'].max()
greatestTimepoints_df = pd.DataFrame(greatestTimepoints)

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
finalTumorVolume_df = pd.merge(greatestTimepoints_df, mouse_study_reduced_df, on=['Mouse ID', "Timepoint"])

# Check that DataFrame is assembled correctly
# finalTumorVolume_df
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

# Create empty list to fill with tumor vol data (for plotting)
tumorVols = []

# Change DataFrame index to Drug Regimen for tumor volume retrieval
df=finalTumorVolume_df.set_index('Drug Regimen')

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
# While iterating through the DataFrame, begin Generating a box plot that shows
# the distrubution of the tumor volume for each drug regimen 
 
fig1, ax1 = plt.subplots()
outlierFormat = dict(marker='o', markerfacecolor='red', markersize=8, linestyle='none')

# Initialize position counter for box plot
i = 1   

for drug in treatments:
    # Locate the rows which contain mice on each drug and get the tumor volumes
    vols = df.loc[[drug],['Tumor Volume (mm3)']]
    tumorVols = vols['Tumor Volume (mm3)'].tolist()
    
    # Add data for current Drug Regimen to box plot
    ax1.boxplot(tumorVols, positions=[i], widths=0.5, flierprops=outlierFormat)
    i += 1
    
    # Determine outliers using upper and lower bounds
    outliers = []
    Q1, Q3 = np.percentile(tumorVols,[25,75])
    IQR = Q3-Q1
    for vol in tumorVols:
        if (vol < Q1-1.5*IQR) | (vol > Q3+1.5*IQR):
            outliers.append(vol)
    print(f"{drug}'s interquartile range is {IQR} and it's potential outliers are: {outliers}.")    

ax1.set_title('Distribution of Final Tumor Volume by Drug Regimen')
ax1.set_ylabel('Tumor volume (mm3)')
ax1.set_xticklabels(treatments)
ax1.set_xlabel('Drug Regimen')
plt.show()
```


## Line and Scatter Plots


```python
# Generate a line plot of tumor volume vs. time point for a single mouse treated with Capomulin
Capomulindf = mouse_study_reduced_df.loc[mouse_study_reduced_df['Drug Regimen']=='Capomulin',:]
capMiceArray = Capomulindf['Mouse ID'].unique()

# Choose randomly a mouse from the array of mice treated with Capomulin
mouseIndex = random.randint(1, len(capMiceArray))
mouseID = capMiceArray[mouseIndex]

# Create a DataFrame for chosen mouse to plot line
mouseDF = Capomulindf.loc[Capomulindf['Mouse ID']==mouseID, ['Timepoint', 'Tumor Volume (mm3)']]

ax = mouseDF.plot(x='Timepoint', y='Tumor Volume (mm3)', kind='line', legend=False)
plt.xlabel('Timepoint (days)')
plt.ylabel('Tumor Volume (mm3)')
plt.title(f'Capolumin Treatment of Mouse number {mouseID}')

# Set y limits based on values close to but outside the whiskers on the box plot for consistency
# when viewing different mice during analysis
ax.set_ylim(20, 50)
plt.show()
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen
xVals = []   # Weight list
yVals = []   # Average Tumor Volume list
for mouse in capMiceArray:
    weight = Capomulindf.loc[Capomulindf['Mouse ID']==mouse, 'Weight (g)'].mean()
    averageTumorVol = Capomulindf.loc[Capomulindf['Mouse ID']==mouse, 'Tumor Volume (mm3)'].mean()
    xVals.append(weight)
    yVals.append(averageTumorVol)
plt.scatter(xVals, yVals, marker='o', facecolors='navy', alpha=.75)
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.title('Mouse Weight vs. Average Observed Tumor Volume')
plt.show()
```


## Correlation and Regression


```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen

# From the scatter plot above, xVals = weights, and yVals = Average Tumor Volumes
correlation = st.pearsonr(xVals, yVals)
print(f"The correlation between mouse weight and average tumor volume is {round(correlation[0],2)}")

xValsArray = np.array(xVals)
yValsArray = np.array(yVals)
(slope, intercept, rvalue, pvalue, stderr) = linregress(xVals, yVals)

regress_values = xValsArray * slope + intercept

line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(xVals, yVals, marker='o', facecolors='navy', alpha=.75)
plt.plot(xVals, regress_values, "r-")
plt.annotate(line_eq,(5.8, 0.8), fontsize=15, color='red')
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.title('Mouse Weight vs. Average Observed Tumor Volume')
plt.show()
```
